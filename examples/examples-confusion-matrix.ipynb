## dummy data for examples

```python
!pip install xgboost
```

```python
from xgboost import XGBClassifier

np.random.seed(100)

X, y = make_classification(
    n_samples=10000, 
    n_features=25,
    n_informative=10,
    n_redundant=0,
    n_classes=5)

class_names = ['class-1','class-2','class-3','class-4','class-5']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)

clf = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')
clf.fit(X_train,y_train)

y_pred = clf.predict(X_test)
```

## simplest possible example, default options

```python
plotting.plot_confusion_matrix(y_test,y_pred,class_names)
```

## add color bar

```python
plotting.plot_confusion_matrix(y_test,y_pred,class_names,colorbar=True)
```

## alternate color scheme

```python
plotting.plot_confusion_matrix(y_test,y_pred,class_names,cmap=plt.cm.Oranges)
```
